```python
#Usecase on Regression Prblem -->derive relationship between your dependent and
#independent variables
data = pd.read_csv('Advertising.csv')
data
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
#plt.hist(data['sales'])
#plt.show()
```

```python
data.columns
```

```python
#Dropping unnecessary column
data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
data.columns
```

```python
#Divide our data into dependent and independent variables
X = data['TV'].values
type(X)
```

```python
Y = data['sales'].values
type(Y)
```

```python
#let's start calculating mean
#calculating the coefficient mean of x and y
#for finding mean value
mean_x = np.mean(X)
print(mean_x)
mean_y = np.mean(Y)
print(mean_y)
```

```python
numer = 0
numer += (X[0] -mean_x)*(Y[0]-mean_y)
numer
denom =0
denom += (X[0]-mean_x)**2
f = numer/denom
f
```

```python
#y = b0 + b1.X b1 -->m nd b0 --> c

n = len(X)
#using the formula to calculate b1(m) and b0(c)
numer = 0
denom = 0
for i in range(n):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2
b1 = numer/denom
print(b1)
b0 = mean_y - (b1*mean_x)
print(b0)
```

```python
print("The linear model is : Y = {:.5}+{:.4}X".format(b0,b1))
```

```python
import matplotlib.pyplot as plt
```

```python
#plotting the values and regression line
max_x = np.max(X) + 100
min_x = np.min(X) -100
#calculating the line values  and y
x = np.linspace(min_x,max_x,1000) #dividing evenly spaced values 
#within gven interval
y = b0 + b1*x
#plottling line -->new values
plt.plot(x,y,color='red',label='Regressionline')
#plotting scatter points -->for our input data
plt.scatter(X,Y,color="blue",label="Scatter Plot")
plt.xlabel('Money spent on TV ads ($)')
plt.ylabel('Sales($)')
plt.title("Basic MAthematical Regression scenario")
plt.legend(loc='best')
plt.show()
```

```python
y = b1*X[10]+b0
y
```

```python
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = b0 + b1*X[i]
    ss_tot += (Y[i]-mean_y)**2
    ss_res += (Y[i]-y_pred)**2
r2 = 1 - (ss_res / ss_tot)
print(r2)
print("R-Squared value is",r2)   
```
